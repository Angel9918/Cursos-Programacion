Introduccion a Matplotlib

```python
import numpy as np
import pandas as pd

file = "Canada.csv"
df_canada = pd.read_csv(file)
df_canada.drop(['Unnamed: 43', 'Unnamed: 44', 'Unnamed: 45', 'Unnamed: 46', 'Unnamed: 47', 'Unnamed: 48', 'Unnamed: 49', 'Unnamed: 50'],axis=1,inplace=True)
df_canada.drop(df_canada.index[195:1015],axis=0,inplace=True)
df_canada.to_csv('df_clean.csv')
```

```python
# df_canada.info(verbose=False)
# df_canada.columns.tolist()
# df_canada.values.tolist()
# df_canada.shape
df_canada.info()
```

```python
# df_canada.drop(['AREA','REG','DEV','Type','Coverage'],axis=1,inplace=True)
df_canada.rename(columns={'OdName':'Country','AreaName':'Continent','RegName':'Region'},inplace=True)
df_canada.columns
```

```python
# Agregar una nueva columna calculada
df_canada['Total']=df_canada.sum(axis=1)
df_canada.head(2)
```

```python
df_canada.describe()
```

Indexacion y seleccion

```python
# Ajusta la columna de paises para ser los indices
df_canada.set_index('Country',inplace=True)

# Borra el nombre de la columna de indices
df_canada.index.name = None
```

```python
df_canada.loc['Japan']
df_canada.iloc[87]
df_canada[df_canada.index == 'Japan']
```

```python
# Obtener el valor para japon en el aÃ±o 2013
df_canada.loc['Japan','2013'] # df_canada.loc[87,36]

# Obtener para japon en un rango de aÃ±os
df_canada.loc['Japan',['1980','1981','1982','1983','1984','1985']] # df_canada.loc[87,[3,4,5,6,7,8]]
```

```python
# Cambio de tipo de dato para las columnas
df_canada.columns = list(map(str,df_canada.columns))
years = list(map(str,range(1980,2014)))
```

```python
# Filtrar para contiente Asia
df_canada.loc[df_canada['Continent']=='Asia']

#Filtrar para continente asia y region southern asia
df_canada.loc[(df_canada['Continent']=='Asia')&(df_canada['Region']=='Southern Asia')]
```

Visualizacion con Matplotlib

```python
import matplotlib as mpl
import matplotlib.pyplot as plt

haiti = df_canada.loc['Haiti',years]
haiti.plot(kind='line')
```

```python
haiti.index = haiti.index.map(int)
haiti.plot(kind='line')
plt.title('Inmigracion desde Haiti')
plt.ylabel('Numero de inmigrantes')
plt.xlabel('AÃ±os')

plt.text(2000,6000,'Terremoto 2010') # Coordenas x,y y texto

plt.show()
```

```python
# Pandas grafica los indices en el eje X y las columnas como lineas individuales en el eje y

China_India = df_canada.loc[['India','China'],years]
China_India = China_India.transpose()

# Cambiar el tipo de datos  de los indices dado
China_India.index = China_India.index.map(int)
China_India.plot(kind='line')

plt.title('Migraciones desde India y China')
plt.xlabel('Cantidad de migrantes')
plt.ylabel('AÃ±os')
plt.show()
```

```python
# Se ordenan los datos por total descendente
df_canada.sort_values(by='Total',ascending=False,axis=0,inplace=True)

# Se obtiene los primeros 5 paises
df_top5 = df_canada.head(5)

# Se transponen los datos por la columna aÃ±os
df_top5 = df_top5[years].transpose()

# Se cambia el tipo de dato de los aÃ±os a entero
df_top5.index = df_top5.index.map(int)

# Se configura el grafico
df_top5.plot(kind='line')

plt.title('5 paises con mayor cantidad de migraciones')
plt.xlabel('Numero de migrantes')
plt.ylabel('AÃ±os')
plt.show()
```
