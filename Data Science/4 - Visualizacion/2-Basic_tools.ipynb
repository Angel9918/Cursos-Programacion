```python
import numpy as np
import pandas as pd

file = "df_clean.csv"
df_canada = pd.read_csv(file)
df_canada
```

```python
df_canada.drop(['Unnamed: 0','Type','Coverage','AREA','REG','DEV'],axis=1,inplace=True)
df_canada.rename(columns={'OdName':'Country','AreaName':'Continent','RegName':'Region'},inplace=True)
df_canada.columns = list(map(str,df_canada.columns))
years = list(map(str,range(1984,2014)))
df_canada['Total'] = df_canada.sum(axis=1)
```

```python
#df_canada.set_index('Country',inplace=True)
df_canada.index.name = None
df_canada
```

```python
df_top5 = df_canada.sort_values(by='Total',ascending=False)
df_top5 = df_top5.head(5)
df_top5
```

```python
df_top5.drop(['Total'],axis=1,inplace=True)
df_top5_T = df_top5[years].transpose()
df_top5_T
```

Area Plot

```python
import matplotlib as mpl
import matplotlib.pyplot as plt


df_top5_T.plot(kind='area',alpha=0.25,stacked=False)
plt.title('Paises con mayor migrantes')
plt.ylabel('Cantidad de inmigrantes')
plt.xlabel('Años')
```

Histograma

```python
count, bin_edges = np.histogram(df_canada['2013'])

df_canada['2013'].plot(kind='hist', xticks=bin_edges)
plt.title('Migracion de 195 paises en 2013')
plt.xlabel('Numero de inmigrantes')
plt.ylabel('Numero de paises')
plt.xticks(rotation=90)
plt.show()
```

Grafico de barras

```python
df_canada.loc['Iceland',years].plot(kind='bar')
plt.title('Inmigracion de Islandia a Canada')
plt.xlabel('Años')
plt.ylabel('Cantidad de inmigrantes')
plt.show()
```
